```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
```

```python
data_canada = px.data.gapminder().query('country == "Canada"')
fig = px.bar(data_canada, x='year', y='pop')
fig
```

```python
data_canada.head()
```

```python
data = px.data.gapminder()

data_canada = data[data.country == 'Canada']
fig = px.bar(
    data_canada,
    x = 'year', y = 'pop',
    hover_data = ['lifeExp', 'gdpPercap'],
    labels = {'pop': 'population of Canada'},
    color = 'lifeExp',
    height = 400
)
fig
```

```python
df = px.data.tips()
fig = px.bar(
    df,
    x='sex', y='total_bill',
    color='time',
    height=400)
fig
```

```python
fig = px.bar(
    df,
    x='sex', y='total_bill',
    color='smoker',
    barmode='group',
    height=400
)
fig
```

```python
fig = px.bar(
    df,
    x='sex', y='total_bill',
    color='smoker',
    barmode='group',
    facet_row='time', facet_col='day',
    category_orders={
        'day': ['Thur', 'Fri', 'Sat', 'Sun'],
        'time': ['Lunch', 'Dinner']
    },
    height=400
)
fig
```

```python
animals = ['giraffes', 'monkeys', 'orangutans']

fig = go.Figure(go.Bar(x=animals, y=[20, 14, 25]))
fig
```

```python
fig = go.Figure(data=[
        go.Bar(name='SF ZOO', x=animals, y=[20, 14, 25]),
        go.Bar(name='LA ZOO', x=animals, y=[12, 18, 29])
])

fig.update_layout(barmode='group')
fig
```

```python
fig = go.Figure(data=[
        go.Bar(name='SF ZOO', x=animals, y=[20, 14, 25]),
        go.Bar(name='LA ZOO', x=animals, y=[12, 18, 29])
])

fig.update_layout(barmode='stack')
fig
```

```python
x = ['Product A', 'Product B', 'Product C']
y = [20, 14, 23]

fig = go.Figure(go.Bar(
    x=x, y=y,
    hovertext=['27% market share', '24% market share', '19% market share']
))

# customize
fig.update_traces(
    marker_color='rgb(158, 202, 225)',
    marker_line_color='rgb(8, 48, 107)',
    marker_line_width=1.5,
    opacity=0.6
)
fig.update_layout(title_text='January 2013 Sales Report')
fig
```

```python
fig = go.Figure(go.Bar(
    x=x, y=y,
    text=y,
    textposition='auto'
))

fig
```
